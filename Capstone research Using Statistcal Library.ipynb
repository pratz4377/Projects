```python
import pandas as pd

df = pd.read_csv('/content/drive/MyDrive/Education/API_4_DS2_en_csv_v2_2807.csv')
```

```python
# Check for missing values in each column
missing_values = df.isnull().sum()
print(missing_values)

```

```python
columns_to_drop = [str(year) for year in range(1960, 2000)]
df = df.drop(columns=columns_to_drop)
```

```python
df
```

```python
# Drop rows with missing values
df = df.dropna()

# Or drop columns with a high percentage of missing values
threshold = 0.5  # Drop columns with more than 50% missing values
df = df.dropna(axis=1, thresh=int(threshold * len(df)))
df
```

```python
df = df.drop_duplicates()
df
```

```python

# Remove extra spaces and set column names to lowercase
df.columns = df.columns.str.strip().str.lower()
df.columns
```

```python
# Convert columns to numeric where applicable
for col in df.columns[4:]:  # Assuming the first four columns are non-numeric
    df[col] = pd.to_numeric(df[col], errors='coerce')
```

```python
df.to_csv("cleaned_data.csv", index=False)
```

```python
selected_countries = ['United States', 'India', 'China', 'Brazil', 'Germany']
```

# **1. Trend Analysis for Selected Indicators Over Time**

```python
import matplotlib.pyplot as plt

# A list of countries and indicators to visualize
indicators = ['Population ages 15-64 (% of total population)',
              'Primary education, duration (years)']

plt.figure(figsize=(10, 5))
for indicator in indicators:
    for country in selected_countries:
        data = df1[(df1['country name'] == country) & (df1['indicator name'] == indicator)]
        years = data.columns[4:]
        values = data.iloc[0, 4:]

        plt.plot(years, values, label=f'{country} - {indicator}')

plt.xlabel('Year')
plt.ylabel('Value')
plt.title('Trend of Selected Indicators Over Time')
plt.xticks(rotation=45)
plt.legend()
plt.show()
```

# **2. Heatmap of single Indicator Across Countries**


```python
import seaborn as sns

# Selecting few indicators for heatmap visualization
selected_indicators = ['Population ages 15-64 (% of total population)']

# Filter data for selected indicators and latest year (2023)
heatmap_data = df[(df['country name'].isin(selected_countries)) &
                  (df['indicator name'].isin(selected_indicators)) &
                  (df['2023'].notna())]
heatmap_data = heatmap_data.pivot(index='country name', columns='indicator name', values='2023')

plt.figure(figsize=(10,5))
sns.heatmap(heatmap_data, annot=True, cmap='coolwarm', cbar_kws={'label': 'Value'})
plt.title('Comparison of Selected Indicators Across Countries (2023)')
plt.xlabel('Indicator')
plt.ylabel('Country')
plt.show()
```

# **3. Comparison of Education Duration Across Countries**


```python
# Filter for education duration indicators
edu_duration = df1[df1['indicator name'].isin(['Primary education, duration (years)',
                                             'Secondary education, duration (years)']) &
                  df1['country name'].isin(selected_countries)]
# Using latest available year for comparison
edu_duration_2023 = edu_duration[['country name', 'indicator name', '2023']].pivot(
    index='country name', columns='indicator name', values='2023')

edu_duration_2023.plot(kind='bar', figsize=(10, 5))
plt.xlabel('Country')
plt.ylabel('Years')
plt.title('Primary and Secondary Education Duration by Country (2023)')
plt.xticks(rotation=90)
plt.legend(title='Education Stage')
plt.show()
```

#**4. Corelation Metrix**

```python
# Corelation metrix of selected indicators according to seleceted couuntries
import seaborn as sns
import matplotlib.pyplot as plt

# Select indicators for correlation analysis
selected_indicators = [
    'Government expenditure on education (% of GDP)',
    'Literacy rate, adult total (% of people ages 15 and above)',
    'Population ages 15-64 (% of total population)',
    'Primary education, duration (years)',
    'Secondary education, duration (years)'
]

# Filter data for selected countries and indicators
correlation_data = df1[(df1['country name'].isin(selected_countries)) &
                      (df1['indicator name'].isin(selected_indicators))]

# Reshape data for correlation
correlation_data = correlation_data.pivot(index='country name', columns='indicator name', values='2023')

# Drop rows with missing values
correlation_data = correlation_data.dropna()

# Compute correlation matrix
correlation_matrix = correlation_data.corr()

# Plot heatmap of correlation matrix
plt.figure(figsize=(10, 5))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', cbar_kws={'label': 'Correlation Coefficient'})
plt.title('Correlation Matrix of Selected Indicators (2023)')
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming df1 is already loaded with the data

# Select the indicators for the scatter plot
indicator_x = 'Population ages 15-64 (% of total population)'
indicator_y = 'Primary education, duration (years)'

# Filter data for the selected indicators and countries
scatter_data = df1[(df1['indicator name'].isin([indicator_x, indicator_y])) &
                   (df1['country name'].isin(selected_countries))]

# Pivot the data to have countries as index and indicators as columns
scatter_data = scatter_data.pivot(index='country name', columns='indicator name', values='2023')

# Create the scatter plot
plt.figure(figsize=(7, 5))
plt.scatter(scatter_data[indicator_x], scatter_data[indicator_y])

# Add labels for each point
for i, country in enumerate(scatter_data.index):
    plt.annotate(country, (scatter_data[indicator_x][i], scatter_data[indicator_y][i]))

plt.xlabel(indicator_x)
plt.ylabel(indicator_y)
plt.title(f'Scatter Plot: {indicator_x} vs {indicator_y} (2023)')
plt.grid(True)
plt.tight_layout()
plt.show()
```
